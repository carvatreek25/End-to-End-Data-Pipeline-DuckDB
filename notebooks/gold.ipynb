```python
import duckdb
```

```python
# Conectar ao DuckDB
conn = duckdb.connect('medalhao.duckdb')
```

```python
# Criar schema gold
conn.execute("CREATE SCHEMA IF NOT EXISTS gold")
```

```python
# Criar tabela de performance por plano
conn.execute("""
    CREATE OR REPLACE TABLE gold.performance_planos AS 
    SELECT 
        tipo_plano,
        status,
        COUNT(id_cliente) as total_clientes,
        SUM(valor_mensalidade) as faturamento_total,
        ROUND(AVG(valor_mensalidade), 2) as ticket_medio,
        CURRENT_DATE as data_extracao
    FROM silver.clientes
    GROUP BY tipo_plano, status
    ORDER BY faturamento_total DESC
""")

print("Camada Gold processada! Dados prontos para análise.")
```
