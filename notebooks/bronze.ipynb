Aqui vamos começar a fazer a ingestão pela Camada Bronze

```python
import duckdb
```

```python
# Conectar ao DuckDB
conn = duckdb.connect('medalhao.duckdb')
```

```python
# Criar schema bronze
conn.execute("CREATE SCHEMA IF NOT EXISTS bronze")
```

```python
# Carregar CSV diretamente para tabela bronze
conn.execute("""
    CREATE TABLE IF NOT EXISTS bronze.clientes AS
    SELECT * FROM read_csv_auto('clientes_academia.csv')
""")
```

```python
# Verificar dados carregados
resultado = conn.execute("SELECT COUNT(*) as total FROM bronze.clientes").fetchall()
print(f"Total de registros carregados: {resultado[0][0]}")
```

```python
# Ver estrutura
conn.execute("DESCRIBE bronze.clientes").df()
```
