Aqui vamos criar a camada Silver e fazer algumas transformações

```python
import duckdb
```

```python
# Conectar ao DuckDB
conn = duckdb.connect('medalhao.duckdb')
```

```python
# Criar schema silver
conn.execute("CREATE SCHEMA IF NOT EXISTS silver")
```

```python


# Transformação completa dos dados da academia
conn.execute("""
    CREATE OR REPLACE TABLE silver.clientes AS 
    SELECT 
        CAST(id_cliente AS INTEGER) as id_cliente,
        UPPER(TRIM(nome)) as nome,
        LOWER(TRIM(email)) as email,
        CAST(data_nascimento AS DATE) as data_nascimento,
        UPPER(plano) as tipo_plano,
        status,
        CAST(data_adesao AS DATE) as data_adesao,
        CAST(valor_mensalidade AS DECIMAL(10,2)) as valor_mensalidade,
        CURRENT_TIMESTAMP as data_processamento
    FROM bronze.clientes
    WHERE email IS NOT NULL 
      AND email LIKE '%@%'
""")

print("Camada Silver populada com sucesso!")
```
